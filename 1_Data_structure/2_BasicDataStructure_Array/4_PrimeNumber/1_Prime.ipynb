```python
# enumerate the prime number that under 1000

cnt = 0

for n in range(2, 1001):
    for i in range(2,n):
        cnt += 1
        if n % i == 0:
            break
        
    else: 
        print(n)
print(f'number of divisions: {cnt}')
```

```python
# improved 1
cnt = 0 # counting about number of division
ptr = 0 # number of already found the prime number 
prime = [None]*500 # array that store the prime number

prime[ptr] = 2 # 2 is prime number
ptr += 1

for n in range(3, 1001, 2):
    for i in range(1, ptr):
        cnt += 1
        if n % prime[i] == 0:
            break
    else: 
        prime[ptr] = n
        ptr += 1
        
for i in range(ptr):
    print(prime[i])
print(f'number of operated division: {cnt}')
```
